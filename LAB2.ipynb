```python
import pandas as pd
import numpy as np
```

```python
file_path = r'C:\Users\User\Downloads\SEM_4_MACHINE_LEARNING\lab_session_data.xlsx'
```

```python
df = pd.read_excel(file_path)
```

```python
# Display the first 5 rows
print(df.head())

# Display information about the DataFrame (data types, non-null values)
print(df.info())
```

```python
# Feature matrix X
X = df[["Candies (#)", "Mangoes (Kg)", "Milk Packets (#)"]].values

# Output vector y
y = df["Payment (Rs)"].values

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
#shape
X.shape[0]
```

```python
#To calculate the rank
rank_X = np.linalg.matrix_rank(X)
print("Rank of feature matrix X:", rank_X)
```

```python
# Pseudo-inverse of X
X_pinv = np.linalg.pinv(X)

# Cost vector
costs = X_pinv @ y

print("Cost of Candies (Rs/unit):", costs[0])
print("Cost of Mangoes (Rs/kg):", costs[1])
print("Cost of Milk Packets (Rs/unit):", costs[2])
```
