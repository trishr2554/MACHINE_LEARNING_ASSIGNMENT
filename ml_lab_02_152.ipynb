A5 SIMILARITY MEASURE(JACCARD AND SMC)

```python
#A5.1 SELECT FIRST TWO OBSERVATION VECTORS
data_two_val = thyroid.iloc[0:2]
data_two_val
```

```python
#A5.2 SELECT ONLY BINARY ATTRIBUTES
binary_attributes = []

for col in data_two_val.columns:
    unique_vals = data_two_val[col].dropna().unique()
    if set(unique_vals).issubset({0, 1}):
        binary_cols.append(col)

binary_attributes
```

```python
#A5.3 CREATE BINARY VECTORS 
binary_data_vector = data_two_val[binary_attributes]
binary_data_vector
```

Now we have two binary vectors:

Vector A → first row

Vector B → second row

```python
#A5.4 COUNT F11,F10,F01,F00  COZ BINARY
def count_binary(a, b):
    f11 = f10 = f01 = f00 = 0
    
    for i in range(len(a)):
        if a[i] == 1 and b[i] == 1:
            f11 += 1
        elif a[i] == 1 and b[i] == 0:
            f10 += 1
        elif a[i] == 0 and b[i] == 1:
            f01 += 1
        else:
            f00 += 1
            
    return f11, f10, f01, f00
```

```python
#A5.5 APPLY THE FUNCTION
A = binary_data_vector.iloc[0].values
B = binary_data_vector.iloc[1].values

f11, f10, f01, f00 = count_binary(A, B)
f11, f10, f01, f00
```

```python
#A5.6 COMPUTE JACCARD COEFIECIENT (JC)
if (f11 + f10 + f01) == 0:
    JC = 0
else:
    JC = f11 / (f11 + f10 + f01)

JC
```

```python
#A5.7 COMPUTE SIMPLE MATCHING COEFFICIENT (SMC)
total_features = f11 + f10 + f01 + f00

if total_features == 0:
    SMC = 0
else:
    SMC = (f11 + f00) / total_features

SMC
```

#A5.8 COMPARE JC AND SMC:
Jaccard coefficient considers only positive matches (1,1),
while Simple Matching Coefficient considers both positive and negative matches.
SMC is more suitable when both 0s and 1s are equally important.

A6 COSINE SIMILARITY MEARURE

```python
#A6.1 SELECT FIRST 2 OBSERVATION VECTORS 
two_vector_data = thyroid.iloc[0:2]
two_vector_data
```

```python
#A6.2 CONVERT DATA TO NUMERIC FORM SINCE COSINE SIMILARITY ONLY WORKS ON NUMBERS
vec_numeric = two_vector_data.select_dtypes(include="number")
vec_numeric
```

```python
#A6.3 SELECT THE VECTORS
A = vec_numeric.iloc[0].values
B = vec_numeric.iloc[1].values
```

```python
import numpy as np

# A6.4 DEFINE COSINE SIMILARITY FUNCTION
def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    if norm_a == 0 or norm_b == 0:
        return 0

    return dot_product / (norm_a * norm_b)

# A6.5 COMPUTE COSINE SIMILARITY
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

cos_similarity= cosine_similarity(a, b)
cos_similarity
```

#A6.6 INTERPRETATION
Cosine similarity measures the angle between two vectors.
A value close to 1 indicates high similarity,
while a value close to 0 indicates low similarity.
